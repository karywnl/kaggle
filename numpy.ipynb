# **NumPy - Numerical Python**

NumPy's ndarray is a fast, ***homogeneous*** multidimensional array. The power is in ***vectorized computation*** doing math on entire arrays without Python loops, which is orders of magnitude faster for large data.


```python
import numpy as np
```

```python
data = np.array([[1, 3, 0.5],
                        [9, 10, 3]])
data
```

```python
data.dtype
```

Note : Most common datatype of choice in numpy array is float64. Just because it is saver and operations such as division always return float.

```python
type(data)
```

```python
np.zeros(10)
```
