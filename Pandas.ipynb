```python
def star_rate(row):
    if row['country'] == 'Canada':
        return 3
    elif row['points'] >= 95:
        return 3
    elif row['points'] >= 85:
        return 2
    else:
        return 1

row = reviews[['country', 'points']].iloc[65]
print(row)
print(star_rate(row))
```

```python
star_ratings = reviews.apply(star_rate, axis='columns')
print(star_ratings)
```

.map(): This is a Series method. It’s a specialist. It transforms each individual element in a single column based on a dictionary, a Series, or a function. It doesn't know other columns exist; it only sees the value in front of it. 🎯
 ***It takes input as dict, series, function***

.apply(): It can work on a Series (element-wise) or a DataFrame. When used on a DataFrame, it can look at entire rows or columns at once. This allows it to perform calculations that involve multiple features (e.g., Total = Price * Quantity). 🛠️
***It takes only function input***

```python
import pandas as pd

df = pd.DataFrame({'Student': ['Miguel', 'Robby', 'Hawk']})
status_map = {'Miguel': 'Champion', 'Robby': 'Finalist', 'Hawk': 'Wildcard'}

# Use .map() to quickly swap names for titles using a dict
df['Status'] = df['Student'].map(status_map)
```

```python
df = pd.DataFrame({
    'Hero': ['Cap', 'Spidey', 'Iron Man'],
    'Strength': [100, 80, 70],
    'Agility': [85, 120, 50]
})

# Use .apply() with axis=1 to look at the whole "row" (all columns)
def calc_hero_score(row):
    return (row['Strength'] * 1.5) + row['Agility']

df['Hero_Score'] = df.apply(calc_hero_score, axis=1)
```

What is the best wine I can buy for a given amount of money? Create a `Series` whose index is wine prices and whose values is the maximum number of points a wine costing that much was given in a review. Sort the values by price, ascending (so that `4.0` dollars is at the top and `3300.0` dollars is at the bottom).

3.
What are the minimum and maximum prices for each variety of wine? Create a DataFrame whose index is the variety category from the dataset and whose values are the min and max values thereof.

```python
price_extremes = reviews.groupby("variety")['price'].agg(['min', 'max'])
```

4.
What are the most expensive wine varieties? Create a variable sorted_varieties containing a copy of the dataframe from the previous question where varieties are sorted in descending order based on minimum price, then on maximum price (to break ties).

```python
sorted_varieties = price_extremes.sort_values(['min', 'max'], ascending=False)

sorted_varieties
```

5.
Create a Series whose index is reviewers and whose values is the average review score given out by that reviewer. Hint: you will need the taster_name and points columns.

```python
reviews.groupby("taster_name")["points"].mean()
```

## 6.
What combination of countries and varieties are most common? Create a `Series` whose index is a `MultiIndex`of `{country, variety}` pairs. For example, a pinot noir produced in the US should map to `{"US", "Pinot Noir"}`. Sort the values in the `Series` in descending order based on wine count.

```python
country_variety_counts = reviews.groupby(["country", "variety"])["winery"].count().sort_values(ascending=False)

country_variety_counts
```
